### Parse SPbU (real parse)

```python
def make_delay():
    time.sleep(1)
```

```python
def make_request_and_save_result(url, path):
    subprocess.run(["wget", "-O", path, url])
```

```python
base_search_url = "https://diploma.spbu.ru/gp/index?GpSearch%5Bname_ru%5D=&GpSearch%5Btitle_ru%5D=&GpSearch%5Beditor_ru%5D=&GpSearch%5Bdp_id%5D=&GpSearch%5Bstatus%5D=1&GpSearch%5Byear%5D=2023&page="
base_artifacts_path = "../artifacts/parsing/spbu"
for page in trange(1, 118, desc="Iterating through search pages"):
    make_delay()
    url = f"{base_search_url}{page}"
    path = f"{base_artifacts_path}/{page}.html"
    make_request_and_save_result(url, path)
```

```python
ids = []
for page in trange(1, 118, desc="Iterating through search pages"):
    path = f"{base_artifacts_path}/{page}.html"
    with open(path, "r") as f:
        soup = BeautifulSoup(f.read(), "html.parser")
    for x in soup.find_all("a", href=True):
        if "id=" in x["href"]:
            ids.append(x["href"].split('=')[-1])
with open(f"{base_artifacts_path}/ids.json", "w") as f:
    json.dump(ids, f)
```

```python
base_view_id_url = "https://diploma.spbu.ru//gp/view?id="
for id in tqdm(ids, desc="Iterating through ids"):
    make_delay()
    url = f"{base_view_id_url}{id}"
    dir_path = Path(f"{base_artifacts_path}/view_id/")
    dir_path.mkdir(exist_ok=True, parents=True)
    path = f"{dir_path}/{id}"
    make_request_and_save_result(url, path)
```
