```python
def get_briefly_summaries(url):
    print("Sleeping before performing request...", end="")
    time.sleep(0.5)
    clear_output()
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser")
    dict_soup = convert(soup)
    plot = dict_soup["html"][0]["head"][0]["meta"][4]["@content"]
    very_short_all_text = dict_soup["html"][0]["body"][0]["main"][0]["article"][0]["div"][1]["#text"]
    data = very_short_all_text.replace("\xa0", " ")
    data = data.replace('\xad', '')
    data = data.replace('\u00ad', '')
    data = data.replace('\N{SOFT HYPHEN}', '')
    if data.startswith("Очень краткое содержание "):
        pos_begin = len("Очень краткое содержание ")
    else:
        pos_begin = 0
    pos_end = data.find("Подробный пересказ по частям")
    data = data[pos_begin:pos_end]
    return plot, data
```

```python
plot, very_short = get_briefly_summaries("https://briefly.ru/tolstoi/karenina/")
```
